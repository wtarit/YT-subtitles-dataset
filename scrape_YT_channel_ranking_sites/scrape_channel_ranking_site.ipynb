```python
import requests
from bs4 import BeautifulSoup
from tqdm.auto import tqdm
```

# https://channelranking.com/ranking-subscribers

```python
r = requests.get("https://channelranking.com/ranking-subscribers")
soup = BeautifulSoup(r.text, features='html.parser')
```

```python
detail_page_urls = []
for link in soup.find_all('a', href=True):
    if '/channel/' in link['href']:
        detail_page_urls.append(link['href'])

detail_page_urls = set(detail_page_urls)
print(len(detail_page_urls))
```

```python
baseurl = "https://channelranking.com"
channel_urls = []
failed_urls = []

for detail_page_url in tqdm(detail_page_urls):
    try:
        full_url = "".join([baseurl, detail_page_url])
        r = requests.get(full_url)
        soup = BeautifulSoup(r.text, features="html.parser")
        div = soup.find("div", style="position: absolute;padding:5px;")
        link = div.find("a", href=True)["href"]
        channel_urls.append(link)
    except Exception as e:
        print(e)
        print(f"FAILED_URL = {detail_page_url}")
        failed_urls.append(detail_page_url)
```

# channelcrawler.com

```python
from pathlib import Path
channelcrawler_html = list(Path("ytcrawler").glob("*.html"))
```

```python
channelcrawler_html
```

```python
channelcrawler_html[0]
```

```python
channelcrawler_yturls = []
for html_file in channelcrawler_html:
    print(html_file)
    with open(html_file, "r", encoding="utf-8") as f:
        html_text = f.read()
    soup = BeautifulSoup(html_text)
    for link in soup.find_all('a', href=True):
        if 'https://youtube.com/' in link['href']:
            channelcrawler_yturls.append(link['href'])

channelcrawler_yturls = set(channelcrawler_yturls)
```

```python
len((channelcrawler_yturls))
```

```python
channelcrawler_yturls
```

```python
with open("channelcrawler_list.txt", "w") as f:
    for item in channelcrawler_yturls:
        f.write(item + "\n")
```
